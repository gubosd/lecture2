## Iris 데이터 속성 분석


```python
# plot graphs in this browser
%matplotlib inline

# prevent unnessesary warnings
import warnings
warnings.filterwarnings('ignore')
#warnings.simplefilter('ignore')
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

- sklearn 이 기본적으로 제공하는 Iris 데이터를 불러온다

```python
from sklearn.datasets import load_iris

iris = load_iris()
print(iris.keys())

X, y = iris.data, iris.target

print('X =',X.shape)
print('y =',y.shape)
```

- 학습용 데이터와 테스트용 데이터로 분리한다

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y) # default -> train 75%, test 25%

print('X_train =',X_train.shape)
print('y_train =',y_train.shape)
print('X_test =',X_test.shape)
print('y_test =',y_test.shape)
```

- 두가지 속성에 대해 산점도를 그려서 개략적인 데이터 형태를 파악한다

```python
import warnings
warnings.filterwarnings('ignore')
#warnings.simplefilter('ignore')

plt.title('Iris (Scatter Map)',fontsize=20)
plt.scatter(X[:,0],X[:,1],c=y,s=60,alpha=0.5) # cmap='ocean'
plt.xlabel('sepal length')
plt.ylabel('sepal width')
plt.colorbar()
```

- 모든 속성쌍에 대해 산점도를 그려보자

```python
fig=plt.figure(figsize=[12,8])
fig.suptitle('Iris\n(setosa:0, versicolor:1, virginica:2)',fontsize=20)
count=0

for i in range(3):
    for j in range(i+1,4):
        count+=1
        plt.subplot(2,3,count)
        plt.scatter(X[:,i],X[:,j],c=y,s=60,alpha=0.5)
        plt.xlabel(iris.feature_names[i])
        plt.ylabel(iris.feature_names[j])

plt.colorbar()
```
