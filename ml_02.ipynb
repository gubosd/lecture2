## 선형회귀 (Linear Regression)
sklearn.linear_model.LinearRegression

```python
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

```python
iris = load_iris()
X_train,X_test,y_train,y_test = train_test_split(iris.data,iris.target)
```

- X_train 에서 3번째 속성(petal length)를 입력값, 4번째 속성(petal width)을 출력값으로 사용
- LinearRegression().fit() 함수를 사용하여 선형회귀 모델 생성

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train[:,2].reshape(-1,1), X_train[:,3]) # reshape() 함수에 주의

w = model.coef_[0] # 기울기
b = model.intercept_ # y 절편

print('w =',w)
print('b =',b)
```

- 결과값인 기울기와 y절편으로 학습한 직선을 그린다 

```python
fig=plt.figure(figsize=(10,8))
plt.title('Linear Regression\n(petal_length vs petal_width)')

plt.scatter(X_train[:,2],X_train[:,3],c=y_train,s=60)
plt.xlabel(iris.feature_names[2])
plt.ylabel(iris.feature_names[3])
plt.colorbar()

plt.plot([0,8],[0*w+b,8*w+b],'y',lw=10,alpha=0.5)
plt.text(0,3,'coef: %f\nintercept: %f' % (w,b), va='top', fontsize=15,color='b')
```

- model.predict() 함수를 사용하여, 테스트 데이타인 X_test 에 대한 예측값을 도출

```python
pred = model.predict(X_test[:,2].reshape(-1,1))
for item in zip(pred,X_test[:,3]):
    print(item)
```
